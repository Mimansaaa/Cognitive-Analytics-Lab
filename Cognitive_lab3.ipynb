## **MEDICAL**


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
medical_df = pd.read_csv('StatewiseTestingDetails.csv', parse_dates=['Date'], dayfirst=True)
```

```python
# Display the first few rows of the dataset
print(medical_df.head())
```

```python
# Display the last few rows of the dataset
print(medical_df.tail())
```

```python
## Data Preprocessing
# Check for missing values
print(medical_df.isnull().sum())

# Handle missing values
df = medical_df.dropna()
```

### **Data Visualization**

```python
# Line Plot of TotalSamples, Negative, and Positive over Time
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['TotalSamples'], label='Total Samples', marker='o')
plt.plot(df['Date'], df['Negative'], label='Negative', marker='o')
plt.plot(df['Date'], df['Positive'], label='Positive', marker='o')
plt.title('COVID-19 Testing Trends over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Bar Plot of Positive Cases by State
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Positive', data=df, palette='muted')
plt.title('Positive Cases by State')
plt.xlabel('State')
plt.ylabel('Positive Cases')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar Plot of TotalSamples by State
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='TotalSamples', data=df, palette='viridis')
plt.title('Total Samples by State')
plt.xlabel('State')
plt.ylabel('Total Samples')
plt.xticks(rotation=90)
plt.show()
```

```python
# Pair Plot for TotalSamples, Negative, and Positive
sns.pairplot(df[['TotalSamples', 'Negative', 'Positive']])
plt.suptitle('Pair Plot for TotalSamples, Negative, and Positive', y=1.02)
plt.show()
```

```python
# Heatmap of Correlation Matrix
correlation_matrix = df[['TotalSamples', 'Negative', 'Positive']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

## **WEATHER**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = '/content/weather.csv'
weather_df = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print(weather_df.head())
```

```python
## Data Preprocessing
# Check for missing values
print(weather_df.isnull().sum())

# Handle missing values
df = weather_df.dropna()
```

### **Data Visualization**

```python
# Scatter Plot of MinTemp vs. MaxTemp
plt.scatter(df['MinTemp'], df['MaxTemp'])
plt.title('Scatter Plot of MinTemp vs. MaxTemp')
plt.xlabel('MinTemp')
plt.ylabel('MaxTemp')
plt.show()
```

```python
# Histogram of WindGustSpeed
plt.hist(df['WindGustSpeed'], bins=20, edgecolor='black')
plt.title('Histogram of WindGustSpeed')
plt.xlabel('WindGustSpeed')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar Plot of WindDir3pm
plt.bar(df['WindDir3pm'], df['WindSpeed3pm'])
plt.title('Bar Plot of WindSpeed3pm by WindDir3pm')
plt.xlabel('WindDir3pm')
plt.ylabel('WindSpeed3pm')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap of Correlation Matrix
correlation_matrix = df.corr()
plt.figure(figsize=(13, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# Count Plot of RainTomorrow
sns.countplot(x='RainTomorrow', data=df)
plt.title('Count Plot of RainTomorrow')
plt.xlabel('RainTomorrow')
plt.ylabel('Count')
plt.show()
```

## **FINANCE**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
finance_df = pd.read_csv('goldstock.csv', parse_dates=['Date'], dayfirst=True)
finance_df = finance_df.sort_values(by='Date')
```

```python
# Display the first few rows of the dataset
print(finance_df.head())
```

```python
# Display the last few rows of the dataset
print(finance_df.tail())
```

```python
## Data Preprocessing
# Check for missing values
print(finance_df.isnull().sum())

# Handle missing values
df = finance_df.dropna()
```

### **Data Visualization**

```python
# Line Plot of Close Prices over Time
plt.plot(df['Date'], df['Close'], marker='o')
plt.title('Close Prices over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```

```python
# Bar Plot of Volume
plt.bar(df['Date'], df['Volume'])
plt.title('Volume over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()
```

```python
# Scatter Plot of Open Prices vs. Close Prices
plt.scatter(df['Open'], df['Close'])
plt.title('Scatter Plot of Open Prices vs. Close Prices')
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.show()
```

```python
# Area Plot of High and Low Prices
plt.fill_between(df['Date'], df['Low'], df['High'], color='skyblue', alpha=0.4)
plt.title('Area Plot of High and Low Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# Histogram of Close Prices
plt.hist(df['Close'], bins=20, edgecolor='black')
plt.title('Histogram of Close Prices')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.show()
```
